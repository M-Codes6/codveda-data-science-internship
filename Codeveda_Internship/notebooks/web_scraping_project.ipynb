# Web Scraping Quotes from a Website

This project demonstrates how to collect data from a website using web scraping techniques.

We extract quotes, authors, and tags from the website "Quotes to Scrape" using Python libraries such as requests and BeautifulSoup.

This task was completed as part of my Data Science Internship at Codveda Technologies.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Step 1: Importing Libraries

In this step, we import the required Python libraries for web scraping.

- **requests**: Used to send HTTP requests and retrieve the webpage content.
- **BeautifulSoup**: Used to parse the HTML content of the webpage and extract specific elements.
- **pandas**: Used to store the scraped data in a structured format and export it as a CSV file.

## Step 2: Sending a Request to the Website

```python
url = "https://quotes.toscrape.com"

response = requests.get(url)

print(response.status_code)
```

## Step 3: Parsing the HTML Content

After successfully sending a request to the website, we receive the webpage content in HTML format.

To extract specific information from the webpage, we use **BeautifulSoup** to parse the HTML and create a structured object that allows us to navigate and search elements easily.

```python
soup = BeautifulSoup(response.text, "html.parser")

print(soup.prettify()[:1000])
```

## Step 4: Extracting Quote Blocks

Each quote on the webpage is stored inside a `<div>` element with the class `quote`.

We use BeautifulSoup to find all these elements and store them for further extraction.

```python
quotes = soup.find_all("div", class_="quote")

print(len(quotes))
```

## Step 5: Extracting Quote Text and Author

Each quote block contains the quote text and the author name.

The quote text is stored inside a `<span>` element with the class `text`, while the author name is stored inside a `<small>` element with the class `author`.

We loop through each quote block and extract this information.

```python
for quote in quotes:

    text = quote.find("span", class_="text").text
    author = quote.find("small", class_="author").text

    print("Quote:", text)
    print("Author:", author)
    print
```

### Storing Extracted Data

After verifying that the scraper correctly extracts quote text and author names, we store the extracted data into Python lists. This allows us to later convert the data into a structured dataset using pandas.

```python
quotes_list = []
authors_list = []

for quote in quotes:

    text = quote.find("span", class_="text").text
    author = quote.find("small", class_="author").text

    quotes_list.append(text)
    authors_list.append(author)

print("Quotes collected:", len(quotes_list))
```

## Step 6: Creating a Dataset with Pandas

After extracting the quotes and authors from the webpage, we convert the collected data into a pandas DataFrame to organize it in a tabular format.

```python
data = {

    "Quote": quotes_list,
    "Author": authors_list
}

df = pd.DataFrame(data)
df.head()
```

## Step 7: Saving the Dataset

After organizing the scraped data into a pandas DataFrame, we save the dataset as a CSV file so it can be used for further analysis or machine learning tasks.

```python
df.to_csv("../datasets/quotes_dataset.csv", index=False)

print("Dataset saved successfully")
```

## Step 8: Scraping Multiple Pages (Pagination)

The quotes website contains multiple pages of quotes.  
Instead of scraping only the first page, we can automate the process using a loop that visits each page and extracts the quotes.

This technique is called **pagination scraping**.

```python
quotes_list = []
authors_list = []

for page in range(1, 6):

    url = f"http://quotes.toscrape.com/page/{page}/"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    for quote in quotes:

        text = quote.find("span", class_="text").text
        author = quote.find("small", class_="author").text

        quotes_list.append(text)
        authors_list.append(author)

print("Total quotes collected:", len(quotes_list))
```

## Step 9: Creating the Final Dataset

After scraping quotes from multiple pages, we organize the collected data into a structured dataset using pandas.

```python
data = {
    "Quote": quotes_list,
    "Author": authors_list
}

df = pd.DataFrame(data)

df.head()
```

## Step 10: Saving the Dataset

The collected data is saved as a CSV file so it can be used for further analysis or machine learning tasks.

```python
df.to_csv("../datasets/quotes_dataset.csv", index=False)

print("Dataset Saved successfully")
```

The extracted data was then organized into a pandas DataFrame and saved as a CSV file for further analysis.
