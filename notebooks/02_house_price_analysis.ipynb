## Step 2: Fixing the Column Separation

Since the dataset values are separated by spaces rather than commas, we reload the dataset using a whitespace separator (`sep="\s+"`).

```python
df = pd.read_csv("../datasets/house_prediction_dataset.csv", sep="\s+")
df.head()
```

### Observation

The dataset structure is now correct, but the column names appear as numeric values. 
This indicates that the first row of the dataset has been incorrectly interpreted as column headers.

```python
df = pd.read_csv("../datasets/house_prediction_dataset.csv", sep="\s+", header=None)
df.head()
```

### Step 3: Verifying the Dataset Structure

After fixing the separator and header issues, we verify that the dataset is correctly loaded.  
We check the dataset structure, summary statistics, and missing values.

```python
df.info()
df.describe()
df.isnull().sum()
```

## Step 4: Assigning Column Names

Currently, the dataset columns are labeled numerically (0–13).
To make the dataset easier to interpret, we assign meaningful column names based on the Boston Housing dataset documentation.

```python
df.columns = [
"CRIM","ZN","INDUS","CHAS","NOX","RM","AGE",
"DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"
]

df.head()
```

### Feature Description

After assigning column names, it is important to understand what each feature represents in the dataset.


| Feature | Meaning |
|------|------|
| CRIM | Crime rate in the area |
| ZN | Percentage of residential land for large houses |
| INDUS | Business area percentage |
| CHAS | Whether the house is near the Charles River (1 = yes, 0 = no) |
| NOX | Air pollution level (Nitric Oxide concentration) |
| RM | Average number of rooms per house |
| AGE | Percentage of older houses |
| DIS | Distance to employment centers |
| RAD | Access to highways |
| TAX | Property tax rate |
| PTRATIO | Student–teacher ratio in schools |
| B | Demographic index |
| LSTAT | Percentage of lower-income population |
| MEDV | Median house price (target variable) |

### Target Variable

In this dataset, the variable **MEDV** represents the median value of owner-occupied homes.  
This will be the **target variable** for prediction, while the remaining features act as input variables.

## Step 5: Exploring the Target Variable

Before analyzing relationships between features and house prices, we first examine the distribution of the target variable **MEDV**.   (EDA)

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MEDV"], bins=30, kde=True)

plt.title("Distribution of House Prices (MEDV)")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")

plt.savefig("../images/house_price_distribution.png")

plt.show()
```

### Observation

The distribution of house prices (MEDV) is slightly right-skewed.  
Most houses fall in the price range of approximately **18–25**.

The dataset also shows a concentration of values at **50**, which indicates a possible upper limit (ceiling effect) in the recorded house prices.

## Step 6: Correlation Analysis

To understand how different features relate to each other and to the target variable (MEDV), we compute the correlation matrix.

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Matrix of Features")

plt.savefig("../images/correalation_heatmap.png")

plt.show()
```

```python
df.corr()["MEDV"].sort_values(ascending=False)
```

### Correlation Insights

From the correlation analysis, the features **RM** (average number of rooms) and **LSTAT** (percentage of lower-income population) show the strongest relationships with house prices (MEDV).

RM has a strong positive correlation with MEDV, indicating that houses with more rooms tend to have higher prices.

LSTAT has a strong negative correlation with MEDV, suggesting that areas with higher proportions of lower-income populations tend to have lower house prices.

## Step 7: Relationship Between Rooms and House Price

Since RM shows the strongest positive correlation with MEDV, we visualize their relationship using a scatter plot.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["RM"], y=df["MEDV"])

plt.title("Relationship Between Number of Rooms and House Price")
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("Median House Value (MEDV)")

plt.savefig("../images/rm_vs_price.png")

plt.show()
```

### Observation

The scatter plot shows a positive relationship between the number of rooms (RM) and house prices (MEDV). 

As the number of rooms increases, the median house value generally increases as well. However, the relationship is not perfectly linear because other factors also influence house prices.

Some outliers can also be observed where houses with similar numbers of rooms have noticeably different prices.

## Step 8: Relationship Between LSTAT and House Price

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["LSTAT"], y=df["MEDV"])

plt.title("Relationship Between Lower Income Population and House Price")
plt.xlabel("Percentage Of Lower Income Population (LSTAT)")
plt.ylabel("Median House Value {MEDV)")

plt.savefig("../images/lstat_vs_price.png")

plt.show()
```

### Observation

The scatter plot shows a strong negative relationship between **LSTAT** (percentage of lower-income population) and **MEDV** (house price).

As LSTAT increases, the median house value tends to decrease.

The pattern also appears slightly curved rather than perfectly linear, suggesting a non-linear relationship between these variables.

### Observation: Ceiling Effect in MEDV

The scatter plot shows many data points with a MEDV value of 50.  
In the Boston Housing dataset, the target variable MEDV was capped at 50, meaning values above 50 were not recorded.

This creates a ceiling effect where multiple houses appear to have the same maximum value even though their true prices may be higher.

## Step 9: Detecting Outliers Using Boxplots

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["MEDV"])

plt.title("Boxplot Of House Prices (MEDV)")
plt.xlabel("Median House Value")

plt.show
```

### Observation

The boxplot shows the distribution of house prices (MEDV).

The median house price is approximately **22**.  
Most of the data lies between **17 and 25**, representing the interquartile range.

Several points above **35** appear as outliers. However, many values are close to **50**, which reflects the ceiling effect present in the dataset.

## Step 10: Pairwise Relationships Between Important Features

```python
sns.pairplot(df[["RM", "LSTAT", "NOX", "MEDV"]])
plt.show()
```

### Pairplot Insights

The pairplot reveals several relationships between the variables:

- **RM and MEDV** show a strong positive relationship. Houses with more rooms tend to have higher prices.
- **LSTAT and MEDV** show a strong negative relationship, indicating that areas with a higher percentage of lower-income population tend to have lower house prices.
- **NOX and MEDV** show a weaker negative relationship, suggesting that higher pollution levels may slightly reduce house prices.

The plots also show clusters where many houses share similar characteristics, particularly around RM values between 5 and 7 and MEDV values between 15 and 30.

## Step 11: Feature Importance Analysis

```python
corr = df.corr()["MEDV"].sort_values(ascending=False)

plt.figure(figsize=(8,5))
corr.plot(kind="bar")

plt.title("Correlation of Features with House Price (MEDV)")
plt.ylabel("Correlation")

plt.show()
```

### Feature Importance Insights

The bar chart shows the correlation of each feature with the target variable MEDV (house price).

The features with the strongest positive relationship with house price are:

- **RM** (average number of rooms)
- **ZN** (proportion of residential land zoned for large lots)
- **B** (demographic index)

Among all variables, **LSTAT** (percentage of lower-income population) has the strongest negative relationship with house prices.

## Key Insights from Exploratory Data Analysis

From the exploratory data analysis of the Boston Housing dataset, several important insights were observed:

- The feature **RM (average number of rooms)** has a strong positive relationship with house prices. Houses with more rooms generally have higher values.
- The feature **LSTAT (percentage of lower-income population)** has a strong negative relationship with house prices, indicating that areas with higher lower-income population tend to have lower property values.
- Pollution levels (**NOX**) and property tax (**TAX**) show moderate negative relationships with house prices.
-  Most houses in the dataset have prices between **17 and 25**.
- The dataset shows a **ceiling effect at MEDV = 50**, meaning house prices above this value were capped at this value.
