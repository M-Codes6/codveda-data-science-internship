# Customer Segmentation using K-Means Clustering (Unsupervised Learning)

This project groups customers based on their annual income and spending behavior using unsupervised learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../datasets/mall_customers.csv")
df.head()
```

```python
df.info()
df.describe()
```

### Observations

- Dataset contains customer demographic and spending data  
- No missing values  
- Features selected for clustering are Annual Income and Spending Score, as they best represent customer purchasing behavior.

```python
df = df.drop("CustomerID", axis=1)
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

### Elbow Method

The elbow point indicates the optimal number of clusters.  
Here, it is around 5 clusters.

```python
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=200, c='red', label='Centroids')

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.legend()
plt.show()
```

## Insights

- Customers are segmented into 5 distinct groups based on income and spending behavior  
- High income & high spending customers are the most valuable segment  
- High income but low spending customers may require better engagement strategies  
- Low income & high spending customers show impulsive buying behavior  
- Low income & low spending customers contribute least to revenue  

Clustering helps businesses target different customer groups with personalized marketing strategies.

### Cluster Interpretation

Each cluster represents a group of customers with similar purchasing behavior.  
Businesses can use this segmentation to design targeted marketing strategies.

## Conclusion

K-Means clustering successfully segmented customers based on income and spending patterns.

This helps businesses improve marketing strategies and customer targeting.
